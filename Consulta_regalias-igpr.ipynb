```python
def quitarna(df):
    for col in listacol64:
        porcnan = df[col].isna().sum()/df.shape[0]
        if porcnan < 0.1:
            df[col] = df[col].fillna(0)
        elif porcnan >= 0.1:
            df[col] = df[col].fillna(df[col].mean())
            
            
    return df
```
